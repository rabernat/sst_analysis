# Generate a spectrum synthetically #

The isotropic energy spectrum is:
$$ E(K) \propto K^{-\alpha} $$
where $ K^2 = k^2+l^2 $. For an arbitrary tracer $ \theta $, the Fourier decomposition can be written as:
$$ \theta(x,y) = \sum \tilde{\theta} (k,l) $$
Since the Fourier components are imaginary numbers, they can be written as, $\tilde{\theta} (k,l) \equiv r_{k,l}e^{i \phi_{k,l}} $
so from the Parseval's theorem, the energy is:
$$ E(k,l) = Re[\tilde{\theta}^*\tilde{\theta}] \approx {r_{k,l}}^2 $$
where the superscript $ * $ represents the conjugate. Taking the azimuthal average yields:
$$ E(K) \approx r^2 \propto K^{-\alpha} $$
and the average energy is given as:
$$ \overline{E} = \int E(k) dk $$
Hence, if we want a spectral slope of $ -5/3 $, i.e. $ \alpha = 5/3 $, the amplitude of the isotropic Fourier components should scale as $ r \propto K^{-5/6} = (k^2+l^2)^{-5/12} $.

For the simulations done by Keating and Smith, the spatial resolution is 1 km and the size of the domain is 1024 x 1024. The domain of the wavenumbers are $ \big[ \frac{1}{1024},\frac{1}{2} \big] $ km$ ^{-1} $.

The amplitude of the Fourier components in the $ k, l $ domain are 
$$ r_{k,l} = (k^2+l^2)^{-5/12} $$
for $ k, l  \in \big[ \frac{1}{1024}, \frac{1}{2} \big] $

```python
import numpy as np
from scipy import io
from scipy import signal as sig
from scipy import fftpack as fft
from scipy import stats
import os
from matplotlib import pyplot as plt
%matplotlib inline
```

## Generate Amplitude ##

```python
N = 512
dL = 1.
k = fft.fftshift(fft.fftfreq(N, dL))
l = fft.fftshift(fft.fftfreq(N, dL))

print k.shape, l.shape
```

```python
kk, ll = np.meshgrid(k, l)

print kk.shape, ll.shape
```

```python
r_kl = (kk**2+ll**2)**(-5./12.)
```

```python
plt.figure(figsize=(8,6))
plt.imshow(kk)
plt.colorbar()

plt.figure(figsize=(8,6))
plt.contourf(np.log10(r_kl), 30)
plt.colorbar()
```

## Generate Random Phase ##

```python
a = np.arange(-np.pi, np.pi, 0.01)
phi = np.random.choice(a, (N, N))

print phi.shape
```

```python
plt.figure(figsize=(8,6))
plt.imshow(phi)
plt.colorbar()
```

```python
F_theta = r_kl*np.exp(1j*phi)

print F_theta.shape
```

```python
plt.figure(figsize=(8,6))
plt.imshow(np.real(np.log10(F_theta)))
plt.colorbar()

plt.figure(figsize=(8,6))
plt.imshow(np.imag(np.log10(F_theta)))
plt.colorbar()
```

## Derive the Field in the xy plane ##

```python
theta = fft.ifft2(fft.ifftshift(F_theta))

print theta.shape
```

```python
x = np.arange(N)
y = np.arange(N)

xx, yy = np.meshgrid(x,y)

print xx.shape, yy.shape
```

```python
plt.figure()
plt.pcolormesh(xx, yy, theta)
plt.colorbar()
```
