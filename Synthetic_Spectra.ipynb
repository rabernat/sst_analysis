# Python FFT algorithm #
$$ y(j) \equiv \sum_{k=0}^{N-1} x(k) e^{-\frac{2\pi i}{n}kj} = \sum_{k=0}^{N-1} x(k) e^{-i\Delta \omega j k} $$
where $ \omega = \Delta \omega j $ and $ \Delta \omega = \frac{2\pi}{n} $. On the other hand, the infinite Fourier transform is
$$ y(j) = \int x(k) e^{-\frac{2\pi i}{n}kj} dk $$
so for the discrete & infinite Fourier transforms to have the same dimensions, 
$$ y(j) = \int x(k) e^{-\frac{2\pi i}{n}kj} dk \approx \Delta k \  \sum_{k=0}^{N-1}  x(k) e^{-\frac{2\pi i}{n}kj} $$
In our case, this becomes
$$ y(k) = \int T(x) e^{-ikx} dx \approx \Delta l \  \sum_{j=0}^{N-1} T(x_j) e^{-ikx_j} = \Delta l \ \mathscr{F} \big( T(x) \big) $$
where $ k $ is the wavenumber and $ x $ is the spatial length. Hence, the PSD is calculated as
$$ s^2(k) = \frac{2}{L} {\Delta x}^2 {\big| \mathscr{F} \big( T(x) \big) \big| }^2 = \frac{2\Delta x}{N} {\big| \mathscr{F} \big( T(x) \big) \big| }^2 $$
with units of $ [^\circ C^2 m] $ and $ L = N \Delta x $.

# Generate a spectrum synthetically #

The isotropic energy spectrum is:
$$ E(K) \propto K^{-\alpha} $$
where $ K^2 = k^2+l^2 $. For an arbitrary tracer $ \theta $, the Fourier decomposition can be written as:
$$ \theta(x,y) = \sum \tilde{\theta} (k,l) $$
Since the Fourier components are imaginary numbers, they can be written as, $\tilde{\theta} (k,l) \equiv r_{k,l}e^{i \phi_{k,l}} $
so from the Parseval's theorem, the energy is:
$$ E(k,l) = Re[\tilde{\theta}^*\tilde{\theta}] \approx {r_{k,l}}^2 $$
where the superscript $ * $ represents the conjugate. Taking the azimuthal average yields:
$$ E(K) \approx r^2 \propto K^{-\alpha} $$
and the average energy is given as:
$$ \overline{E} = \int E(k) dk $$
Hence, if we want a spectral slope of $ -5/3 $, i.e. $ \alpha = 5/3 $, the amplitude of the isotropic Fourier components should scale as $ r \propto K^{-5/6} = (k^2+l^2)^{-5/12} $.

For the simulations done by Keating and Smith, the spatial resolution is 1 km and the size of the domain is 1024 x 1024. The domain of the wavenumbers are $ \big[ \frac{1}{1024},\frac{1}{2} \big] $ km$ ^{-1} $.

The amplitude of the Fourier components in the $ k, l $ domain are 
$$ r_{k,l} = (k^2+l^2)^{-5/12} $$
for $ k, l  \in \big[ \frac{1}{1024}, \frac{1}{2} \big] $

 ## Symmetry of the Fourier Component ##

If the function $ f(x) $ is real, then its Fourier transform:
$$ \tilde{f} (k) = \int \frac{dx}{\sqrt{2\pi}} f (x) e^{-ikx} $$
obeys the relation
$$ {\tilde{f}}^{\ *}(k) = \int \frac{dx}{\sqrt{2\pi}} f (x) e^{ikx} = \tilde{f}(-k) $$

For a 2D case, the Fourier transform is:
$$ \tilde{\theta} (k, l) = \iint \frac{1}{\sqrt{2\pi}} \theta (x, y) e^{-i(kx+ly)} dxdy $$
and obeys the relation
$$ {\tilde{\theta}}^{\ *}(k, l) = \iint \frac{1}{\sqrt{2\pi}} \theta (x, y) e^{i(kx+ly)} dxdy = \tilde{\theta}(-k, -l) $$

Since the Fourier transform is a complex number, $ \tilde{\theta} (k, l) = r(k, l) e^{i \phi (k, l)} $, the complex conjugate can be written as $ {\tilde{\theta}}^{\ *} (k, l) = r(k, l) e^{-i \phi (k, l)} $

When the function $\theta (x, y)$ is real, therefore the relation: 
$$ r(-k, -l) e^{i \phi (-k, -l)} = {\tilde{\theta}} (-k, -l) = {\tilde{\theta}}^{\ *} (k, l) = r(k, l) e^{-i \phi (k, l)} $$
must hold. In other words, if we prescribe the function $ \tilde{\theta} (k, l) $ for $ \textbf{K} = (k, l) $, the Fourier field for $ \textbf{K} = (-k, -l) $ will be the conjugate $ \tilde{\theta}^{\ *} (k, l) $. Hence,
$$ r(-k, -l) = r(k, l), \ \ \ \ \phi(-k, -l) = -\phi(k, l) $$

This means that in order to create a Fourier field that represents a real function in the spatial domain, when the Fourier field is prescribed for $ \textbf{K} = (k, l) $, the amplitude needs to be rotated $180^{\circ}$ and the phase is rotated $180^{\circ}$ with the opposite sign for $ \textbf{K} = (-k, -l) $.

## Structure Function using Module ##

$$ \big\langle {\big|b'(x+ n\Delta x) - b'(x)\big|}^q \big\rangle \propto \big| n\Delta x\big|^{H_q} $$
$$ \big\langle {\big|b'(y+ n\Delta y) - b'(y)\big|}^q \big\rangle \propto \big| n\Delta y\big|^{H_q} $$
where the brackets, $ \langle \rangle $ represent a spatiotemperal averaging and $H_q$ is a constant. The variable $n$ represents the grid spacing.
$$ \therefore \hspace{0.2cm} \Sigma_q^x(n) \equiv \frac {\log{ \big( \big \langle {\big|b'(x+ n\Delta x) - b'(x) \big|}^q \big \rangle \big)}}{\log{ | n\Delta x |}} \propto H_q , \hspace{0.5cm}
\Sigma_q^y(n) \equiv \frac {\log{ \big( \big \langle {\big|b'(y+ n\Delta y) - b'(y) \big|}^q \big \rangle \big)}}{\log{ | n\Delta y |}} \propto H_q $$

```python
p = idealmodel.IDEALFile(fname)
N, Struc_funci, Struc_funcj = p.structure_function(varname='bC')   # Calculates all the variables necessary for the structure function
```

## Fitting Slope to Stucture Function ##

```python
dx = 1e0   # [m]
dy = 1e0
x = 2**np.arange(N)*dx
y = 2**np.arange(N)*dy
X = np.log10(x)
B = np.log10(Struc_funci)

A = np.vstack([X[:2], np.ones(len(X[:2]))]).T
m_1, c = np.linalg.lstsq(A, B[:2])[0]

A = np.vstack([X[1:5], np.ones(len(X[1:5]))]).T
m_2, c = np.linalg.lstsq(A, B[1:5])[0]

print m_1, m_2, c  # slope and intercept
```
