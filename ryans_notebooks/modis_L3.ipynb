```python
import numpy as np
from pyhdf.SD import SD, SDC
import os
```

```python
ddir = '/Volumes/Bucket1/Data/ghrsst/data/L3/aqua/11um/4km/daily/'
fname = os.path.join(ddir, '2008', '001', 'A2008001.L3m_DAY_NSST_4')
a = SD(fname, SDC.READ)
d = a.select('l3m_data')
attr = d.attributes()
sst = attr['Slope']*d[:] + attr['Intercept']
q = a.select('l3m_qual')
qual = q[:]
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
#plt.rcParams['image.origin'] = 'top'
plt.rcParams['figure.figsize'] = (16,8)
```

```python
plt.imshow(sst)
```

```python
plt.imshow(qual)
```
