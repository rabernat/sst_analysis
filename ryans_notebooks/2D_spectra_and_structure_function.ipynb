# Structure Functions and Isotropic Spectra in 2D #

### Webb (1964) / McCaffrey et al. (2014) ###

Webb's original paper deals with only 1D signals and spectra. McCaffrey deals with _isotropic_ spectra. This is a subtle but crucial difference. For the proof of the relationship between spectrum and structure function, the spectrum must satisfy

$$ \langle \theta^2 \rangle = \int_0^\infty B(k_h) dk_h $$

Let's unpack this a little bit.

### 1D Signals ###

For a 1D signal, the expression on the left represents the integral over $x$. For now, we will stick with inifite domains and Fourier transforms, rather than finite periodic domains and Fourier series. The domain-integrated variance is defined as

$$  \langle \theta^2 \rangle = \int_{-\infty}^\infty \theta(x) dx \ . $$

We define the Fourier transform $\hat{\theta}(k)$ via the following duality relations:

$$ \theta(x)       =  \int_{-\infty}^\infty \hat{\theta}(k)e^{i k x} dk \\
   \hat{\theta}(k) =  \frac{1}{2 \pi} \int_{-\infty}^\infty \theta(x)e^{-i k x} dx \ .
$$

The [Plancherel Theorem](http://en.wikipedia.org/wiki/Fourier_transform#Plancherel_theorem_and_Parseval.27s_theorem), easily proved, states that

$$ \int_{-\infty}^\infty | \theta(x) |^2 dx = \int_{-\infty}^\infty | \hat{\theta}(k) | ^2 dk $$

All we have to do to satisfy Webb's condition is to set $k_h = k$, assume that $| \hat{\theta}(k)|^2 = |\hat{\theta}(-k)|^2$ (always true for real-valued $\theta$), and define

$$ B_{1D}(k_h) = 2 |\hat{\theta}(k_h)|^2 \ . $$

### 2D Signals ###

The situation in 2D is more complicated. The domain-integrated variance is

$$  \langle \theta^2 \rangle = \int_{-\infty}^\infty \int_{-\infty}^\infty \theta(x,y) dx dy \ . $$

The Fourier pairs are

$$ \theta(x,y)       =  \int_{-\infty}^\infty \int_{-\infty}^\infty 
                        \hat{\theta}(k,l)e^{i (k x + l y)} dk dl \\
   \hat{\theta}(k,l) =  \frac{1}{(2 \pi)^2} \int_{-\infty}^\infty \int_{-\infty}^\infty 
                        \theta(x)e^{-i (k x + l y)} dx dy \ .
$$

The Plancherel relation still looks the same

$$ \int_{-\infty}^\infty \int_{-\infty}^\infty | \theta(x, y) |^2 dx dy = 
   \int_{-\infty}^\infty \int_{-\infty}^\infty | \hat{\theta}(k,l) | ^2 dk dl \ .$$

However, the isotropic spectrum needs to be defined carefully. For any signal $f(k,l)$, we can transform to polar coordinates $f(k_h, \phi)$, where

$$ k_h = \sqrt{k^2 + l^2} \ , \ \ \ \phi = \arctan(|l/k|) \ . $$

The integral of $f$ over all $k$ and $l$ is written

$$ \int_{-\infty}^\infty \int_{-\infty}^\infty f(k,l) dk dl =
   \int_{0}^\infty \left( \int_{-\pi}^\pi f(k_h, \phi) k_h d\phi \right ) dk_h $$
   
The key point is that, __due to the transformation to polar coordintes, we pick up an extra factor of $k_h$ in the integral__! In order to satisfy Webb's condition in 2D, we therefore have to define

$$ B(k_h) = \int_{-\pi}^\pi | \hat{\theta}(k_h, \phi) | ^2  k_h d\phi $$

This means if we want to have $B(k_h) = B_0 k_h^{-\alpha}$, we need

$$ | \hat{\theta}(k, l) | ^2 = (2 \pi)^{-1} k_h^{-1} B(k_h) = (2 \pi)^{-1} B_0 k_h^{-(\alpha + 1)} $$

```python
def isotropic_synthetic_spectrum(N, slope=3.):
    """Generate a 2D synthetic field of size NxN with an isoptropic spectral slope of -slope."""

    k = np.fft.fftfreq(N)
    l = np.fft.fftfreq(N)
    kk, ll = np.meshgrid(k,l)
    kh = np.sqrt(kk**2 + ll**2)

    # power amplitude
    r = np.ma.masked_invalid(np.sqrt(kh**(-slope-1))).filled(0.)

    # phase
    # just be lazy, take the real part later
    th = 2*np.pi*np.random.rand(N,N) - np.pi

    # complex fourier amplitudes
    a = r * np.exp(1j * th)

    # check that symmetry of FT is satisfied
    #np.testing.assert_almost_equal(a[1:N/2], a[-1:-N/2:-1,::-1].conj())

    f = np.real(np.fft.ifft2(a))
    
    return np.real(f) 
```

```python
def isotropic_power_spectrum(f, nbins=64):
    """Calculate isotropic power spectrum of signal f."""
    
    assert f.ndim == 2
    Ny,Nx = f.shape
    k = np.fft.fftfreq(Nx)
    l = np.fft.fftfreq(Ny)
    kk, ll = np.meshgrid(k,l)
    kh = np.sqrt(kk**2 + ll**2)
    ft = np.fft.fft2(f)
    ps2d = np.real(ft * ft.conj())
    
    Ki = np.linspace(0, k.max(), nbins)
    Ki_c = 0.5 * (Ki[1:] + Ki[:-1])
    Kidx = np.digitize(kh.ravel(), Ki)
    # kh.ravel()[i] < Ki[i]
    
    
    area = np.bincount(Kidx)
    iso_ps = np.ma.masked_invalid(np.bincount(
                Kidx, weights=(ps2d * kh * 2*np.pi).ravel(),
                minlength=nbins+1
            ) / area, )[1:-1] # discard everything below and above the max and min lims
    
    return Ki_c, iso_ps
```

```python
def isotropic_structure_function(f, order=2):
    
    assert f.ndim == 2
    Ny,Nx = f.shape
    # can only deal with square domains for now
    assert Nx == Ny
    N = Nx
    # the length intervals
    # powers of 2
    ns = np.arange(0,np.log2(N/2), dtype='i4')
    l = 2**ns
    
    Ns = len(ns)
    
    # the isotropic structure function
    s = np.zeros(Ns, dtype=f.dtype)
    
    for n in ns:
        dx = np.abs(f - np.roll(f, l[n], axis=0))**order
        dy = np.abs(f - np.roll(f, l[n], axis=1))**order
        s[n] = 0.5*( dx.mean() + dy.mean() )
        
    return l, s
        
```

```python
def fit_loglog(x, y):
    # fig log vs log
    # mask any components where either x or y is zero
    mask = (x<=0.) | (y<=0.)
    p = np.polyfit(np.log2(np.ma.masked_array(x,mask)).compressed(), 
                   np.log2(np.ma.masked_array(y,mask)).compressed(), 1)
    y_fit = 2**(np.log2(x)*p[0] + p[1])
    
    return y_fit, p[0]
    
```

```python
powers_of_two = range(7,12,2)
slopes = [1., 1.5, 2., 2.5, 3.]
```

# Plots and Fits of $F(x)$, $E(k)$, $S(l)$ #

```python
N_ensembles = 8
ps_slopes = np.zeros((len(powers_of_two), len(slopes)))
sf_slopes = np.zeros_like(ps_slopes)

for np2, p2 in enumerate(powers_of_two):

    N=2**p2
    x = np.arange(N)

    for nslope, slope in enumerate(slopes):

        fig = plt.figure(figsize=(16,6))
        ax_ps = plt.subplot2grid((2,N_ensembles),
                                 (1,0), colspan=N_ensembles/2)
        ax_sf = plt.subplot2grid((2,N_ensembles),
                                 (1,N_ensembles/2), colspan=N_ensembles/2)
        
        ps_avg = 0.
        sf_avg = 0.
        for Ne in range(N_ensembles):
            ax = plt.subplot2grid((2,N_ensembles), (0,Ne))
            
            f = isotropic_synthetic_spectrum(N, slope=slope)
            k, ps = isotropic_power_spectrum(f)
            ps_avg += ps
            
            l, sf = isotropic_structure_function(f, order=1)
            sf_avg += sf
            
            ax.imshow(f)
            ax.set_axis_off()
            #ax.set_title('%g x %g' % (N,N))
            
            ax_ps.loglog(k, ps, color='0.5')
            ax_sf.loglog(l, sf, 'o-', color='0.5')

        ps_fit, ps_slope = fit_loglog(k, ps_avg/N_ensembles)
        sf_fit, sf_slope = fit_loglog(l, sf_avg/N_ensembles)
        ps_slopes[np2,nslope] = ps_slope
        sf_slopes[np2,nslope] = sf_slope
        ax_ps.loglog(k, ps_fit, 'm')
        ax_sf.loglog(l, sf_fit, 'm')
            
        #ax_ps.loglog(k, ps_fit,'k-')
        ax_ps.set_xlabel(r'$k_{h}$')
        ax_ps.set_title('Isotropic Power Spectrum - Avg. Slope = %g' % ps_slope)

        #ax_sf.loglog(l, sf_fit, 'k-')
        ax_sf.set_xlabel('l')
        ax_sf.set_title('1st Order S.F. - Avg. Slope = %g' % sf_slope)

        fig.text(0.5, 0.9, 'Slope = %g, Res = %g' % (slope, N), ha='center', fontsize=16)
        #plt.tight_layout()
        plt.show()
```

# Test of Relationship between $H$ and $\alpha$ #

Here we test whether $ \alpha = (1 + 2H) $. We estimate $H$ by doing a log-log fit on the first order structure frunction $S_1(l)$.

```python
plt.figure(figsize=(15,8))

for np2, p2 in enumerate(powers_of_two):

    plt.subplot(2,3,np2+1)

    N=2**p2

    plt.plot( slopes, -ps_slopes[np2], 'o-', label=r'$\alpha$')
    plt.plot( slopes, sf_slopes[np2]*2 + 1, '^-', label=r'$2H + 1$')
    plt.xlim([0.5,3.5])
    plt.xlabel('Predicted')
    plt.ylabel('Diagnosed')
    plt.ylim([0.5,3.5])
    plt.legend(loc='lower right')
    plt.title('Resolution = %g' % N)
    plt.grid()

```
