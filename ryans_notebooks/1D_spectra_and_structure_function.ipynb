# Structure Functions of Synthetic Signals #

In turbulence studies, the typical focus is wavenumber power spectra. (Even though Kolmogorov's original theory was framed in terms of structure functions, not spectral analysis.) Anyway, in the literature, various theories of turbulence predict spectral slopes of the form

$$ E(k) = E_0 k^{-\alpha} \ .$$

Here we generate synthetic signals that satisfy such power laws (with radomized phase), and use them in structure function analysis. These signals are denoted $ F_\alpha(x) $.

## Self Similar Process with Stationary Increments ##

A process $F(x)$ is said to be "self similar with stationary increments" if 

$$ F(x + \Delta) - F(x) \stackrel{d}{=} r^{-H} [ F(x + r\Delta) - F(x) ] $$

where the notation $\stackrel{d}{=}$ means that the two sides have the same probability distribution. In this case, it can be shown that

$$ E(k) \propto k^{-(1+2H)} \ .$$

We define the structure function of $F(x)$ of order $q$ as 

$$ S_q(l) = \langle |F(x + l) - F(x)|^q \rangle $$

where $l$ is a length increment. The definition of the sssi process then gives

$$ S_q(l) \propto l^{Hq} \ . $$

This leads to a relationship between the first order structure function and the spectral slope.

$$ \alpha = (1 + 2H) \ .$$

This scaling is only valid for $0 < H < 1$, corresponding with $1 < \alpha < 3$.

Below we test this relationship with synthetic signals for a range of $\alpha$. We also examine the dependence on $N_x$, the number of samples in the record.


```python
def synthetic_spectrum(N, slope=3.):
    """Generate a synthetic series of length N with a spectral slope of -slope."""

    k = np.fft.fftfreq(N)
    # amplitude
    r = np.ma.masked_invalid(np.abs(k)**(-slope/2.)).filled(0.)
    # phase
    th_half = 2*np.pi*np.random.rand(N/2 - 1) - np.pi
    th = np.zeros_like(r)
    th[1:N/2] = th_half
    th[N/2+1:] = -th_half[::-1]
    # complex fourier amplitudes
    #a = r + 1j*th
    a = r * np.exp(1j * th)
    
    # check that symmetry of FT is satisfied
    np.testing.assert_almost_equal(a[1:N/2], a[-1:-N/2:-1].conj())

    f = np.fft.ifft(a)
    return np.real(f) 
```

```python
def power_spectrum(f):
    """Calculate power spectrum of signal f."""
    
    N = len(f)
    k = np.fft.fftfreq(N)
    ft = np.fft.fft(f)
    ps = np.real(ft * ft.conj())
    return -k[-1:-N/2:-1], ps[-1:-N/2:-1]
```

```python
def structure_function(f, order=2):
    N = len(f)
    # the length intervals
    # powers of 2
    ns = np.arange(0,np.log2(N/2), dtype='i4')
    l = 2**ns
    
    Ns = len(ns)
    
    # the structure function
    s = np.zeros(Ns, dtype=f.dtype)
    # its standard deviation
    s_std = np.zeros_like(s)
    
    for n in ns:
        d = np.abs(f - np.roll(f, l[n]))**order
        s[n] = d.mean()
        
    return l, s
        
```

```python
def fit_loglog(x, y):
    # fig log vs log
    p = np.polyfit(np.log2(x), np.log2(y), 1)
    y_fit = 2**(np.log2(x)*p[0] + p[1])
    
    return y_fit, p[0]
    
```

```python
powers_of_two = range(7,16,2)
slopes = [1., 1.5, 2., 2.5, 3.]
```

# Plots and Fits of $F(x)$, $E(k)$, $S(l)$ #

```python
ps_slopes = np.zeros((len(powers_of_two), len(slopes)))
sf_slopes = np.zeros_like(ps_slopes)

for np2, p2 in enumerate(powers_of_two):

    N=2**p2
    x = np.arange(N)

    for nslope, slope in enumerate(slopes):

        f = synthetic_spectrum(N, slope=slope)

        k, ps = power_spectrum(f)
        ps_fit, ps_slope = fit_loglog(k, ps)
        ps_slopes[np2,nslope] = ps_slope

        l, sf = structure_function(f, order=1)
        sf_fit, sf_slope = fit_loglog(l, sf)
        sf_slopes[np2,nslope] = sf_slope
        
        
        fig = plt.figure(figsize=(16,4))
        
        ax1 = fig.add_subplot(131)
        ax1.plot(x, f)
        ax1.set_xlabel('x')
        ax1.set_title('Original Signal - N = %g' % N)

        ax2 = fig.add_subplot(132)
        ax2.loglog(k, ps)
        ax2.loglog(k, ps_fit)
        ax2.set_xlabel('k')
        ax2.set_title('Power Spectrum - Slope = %g' % ps_slope)

        ax3 = fig.add_subplot(133)
        ax3.loglog(l, sf, 'o-')
        ax3.loglog(l, sf_fit)
        ax3.set_xlabel('l')
        ax3.set_title('1st Order S.F. - Slope = %g' % sf_slope)
        
```

# Test of Relationship between $H$ and $\alpha$ #

Here we test whether $ \alpha = (1 + 2H) $. We estimate $H$ by doing a log-log fit on the first order structure frunction $S_1(l)$.

```python
plt.figure(figsize=(15,8))

for np2, p2 in enumerate(powers_of_two):

    plt.subplot(2,3,np2+1)

    N=2**p2

    plt.plot( slopes, -ps_slopes[np2], 'o-', label=r'$\alpha$')
    plt.plot( slopes, sf_slopes[np2]*2 + 1, '^-', label=r'$2H + 1$')
    plt.xlim([0.5,3.5])
    plt.xlabel('Predicted')
    plt.ylabel('Diagnosed')
    plt.ylim([0.5,3.5])
    plt.legend(loc='lower right')
    plt.title('Resolution = %g' % N)
    plt.grid()

```

### Comments ###

We see that, intedependent of resolution, the relationship holds very well for $H=0.5$ ($\alpha = 2$). But in general, our method of estimating $H$ appears slighly biased, scaling more weakly with $\alpha$ than expected. There is a clear reduction of the bias with resolution. However, the values $\alpha=1$ and $\alpha=3$ correspond with the upper and lower limits of validity of the self-similarity assumptions.

It would be comforting if we could reach a simple explanation for this bias, or even develop a way to correct it. Nevertheless, it is weak enough (especially around $\alpha=2$) that, we should still be able to clearly distinguish different types of turbulence using structure functions.  

# Higher Order Structure Functions #

The difference between a monofractal and multi-fractal has to do with the way the higher order structure functions scale. Below we compute a few higher order structure functions. 

```python
p2 = 15
N=2**p2
x = np.arange(N)

for nslope, slope in enumerate(slopes):
    
    f = synthetic_spectrum(N, slope=slope)

    plt.figure()
    for order in range(1,4):
        l, sf = structure_function(f, order=order)
        sf_fit, sf_slope = fit_loglog(l, sf)
        plt.loglog(l, sf, 'o-', label='order = %1d slope = %3.2f' % (order, sf_slope))
    plt.legend(loc='lower right')
    plt.title('Spectral Slope = %g' % slope)
```
