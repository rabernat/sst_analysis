```python
import numpy as np
from scipy import io
from scipy import signal as sig
from scipy import fftpack as fft
from scipy import stats
import os
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
Nt = (64-25)/5+1
Ny, Nx = f['bC'].shape
Tracerx = np.zeros((Nt, Ny, Nx))
print Tracerx.shape

# load data files
base_dir = 'Phillips/'

nday = 25
index = 0
while nday < 65:
    if nday % 5 == 0:
        fname = os.path.join(base_dir, 
                         'b%02d.mat' % (nday))
        print fname
        #fname = os.path.join(ddir, '2008', '001', 'A2008001.L3m_DAY_NSST_4')
        f = io.loadmat(fname)
        Tracerx[index] = f['bC']
        index += 1
        print index, nday
    nday += 1
```

```python
plt.figure()
plt.imshow(Tracerx[0])

plt.figure()
plt.imshow(Tracerx[7])
```

```python
dx = 1000
dy = 1000

# Hanning window
windowx = sig.hann(Nx)
windowy = sig.hann(Ny)
window = windowx*windowy[:,np.newaxis]

k = fft.fftshift(fft.fftfreq(Nx, dx))
l = fft.fftshift(fft.fftfreq(Ny, dy))

print window.shape, k.shape, l.shape
```

```python
PSD_Bx = np.zeros((Ny,Nx))

for n in range(Nt):
    #for vname, ncvname in varnames.iteritems():
        #mld[n] = nc.variables['HMXL_2'][n].mean()
    Q = Tracerx[n] 
        # remove mean and window
    Q -= Q.mean()
    Q *= window
        # fourier transform
    Qf = fft.fftshift(fft.fft2(Q))
        # isoptropic power spectrum
    PSD_Bx += 2.*np.real(Qf*np.conj(Qf))
        #power_spectrum[vname][n] = (
        #    np.bincount(Ridx, weights=np.real(Qf*np.conj(Qf)).ravel())[:-1] )
    #print n
PSD_Bx = PSD_Bx/Nt
```

```python
plt.pcolormesh(k,l,np.log10(PSD_Bx))
plt.colorbar()
```

```python
kk,ll = np.meshgrid(k,l)
K = np.sqrt(kk**2 + ll**2)
plt.pcolormesh(k, l, K)
```

```python
PSD_Bxx = 2.*dx*PSD_Bx.mean(axis=0)/Nx
PSD_Bxy = 2.*dy*PSD_Bx.mean(axis=1)/Ny
```

```python
p = 0.95

ax, bx = stats.chi2.ppf([(1-p)*0.5, (1+p)*0.5], Nt*Nx)
ay, by = stats.chi2.ppf([(1-p)*0.5, (1+p)*0.5], Nt*Ny)
err_x = 2*Nt*Nx*np.array([1/ax, 1/bx])
err_y = 2*Nt*Ny*np.array([1/ay, 1/by])
```

```python
X = 3e-5
Y = 2e-5
#yJan = 5e0
#yJul = 5e0
A = 1e4

plt.figure(figsize=(10,7))
blue, = plt.loglog(k, PSD_Bxx, basex=10, basey=10)
green, = plt.loglog(l, PSD_Bxy, basex=10, basey=10)
magenta, = plt.loglog([X,X], A*err_x, 'm', basex=10, basey=10)
red, = plt.loglog([Y,Y], A*err_y, 'r', basex=10, basey=10) 
plt.title('SST power spectral in the zonal direction', fontsize=15)
plt.xlabel('wavenumber $(=k/2\pi) [m^{-1}]$', fontsize=15)
plt.ylabel('PSD $[c ^2 m]$', fontsize=15)
#plt.xlim([10**(-7),10**(-3)])
#plt.ylim([10**(-8), 10**1])
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.xticks([10**(-7),10**(-6),10**(-5),10**(-4),10**(-3)],fontsize=15)
#plt.yticks([10**(-7),10**(-6),10**(-5),10**(-4),10**(-3),10**(-2),10**(-1),10**0], fontsize=15)
#plt.legend([blue, green], ['August', 'February'], loc=1, fontsize=15)
#plt.legend([blue, green, magenta, red], ['', 'January', 'Confidence Interval (July)', 'Confidence Interval (January)'], loc=3, fontsize=15)
```

```python
kk, ll = np.meshgrid(k, l)
K = np.sqrt(kk**2 + ll**2)

plt.pcolormesh(k, l, K)
#plt.clim([K[0], K[-1]])
plt.colorbar()
```

```python
def calc_isotropic_spectrum(k, l, spectrum_2d, nbins=128):
    #Ki = k[(Nx/2):]
    kk, ll = np.meshgrid(k, l)
    K = np.sqrt(kk**2 + ll**2)
    Ki = np.linspace(0, k.max(), nbins)

    Kidx = np.digitize(K.ravel(), Ki)
    area = np.bincount(Kidx)
    isotropic_spectrum = np.ma.masked_invalid(
            np.bincount(Kidx, weights=spectrum_2d.ravel()) / area )
    return Ki, isotropic_spectrum[1:], area[1:]
```
