# MODIS Level 2 SST Data #

We are analyzing data from the [MODIS](http://en.wikipedia.org/wiki/Moderate-Resolution_Imaging_Spectroradiometer) (Moderate-Resolution Imaging Spectroradiometer) instrument on the [Aqua](http://aqua.nasa.gov/) satellite platform. More details on this product can be found on the [NASA PODAAC website](http://podaac.jpl.nasa.gov/dataset/JPL-L2P-MODIS_A).

This is ["level 2 data"](http://en.wikipedia.org/wiki/Remote_sensing#Data_processing_levels), which means that the raw sensor data has been processed to give physical variables (SST in our case), but not regridded; the data is reported exactly as it was measured from space. This form of data provides the highest possible spatial resolution, but it is also harder to work with than gridded data.

The data is distributed by the [GHRSST](http://www.ghrsst.org) (Group for High-Resolution SST) in a standardized format. This format is described in a very useful [manual](https://www.ghrsst.org/files/download.php?m=documents&f=111020110451-GHRSSTUserGuidev91.pdf).

The Moderate-resolution Imaging Spectroradiometer (MODIS) is a scientific instrument (radiometer) launched by NASA in 2002 on board the Aqua satellite platform (a second series is on the Terra platform) to study global dynamics of the Earths atmosphere, land and oceans. MODIS captures data in 36 spectral bands ranging in wavelength from 0.4 um to 14.4 um and at varying spatial resolutions (2 bands at 250 m, 5 bands at 500 m and 29 bands at 1 km). For the sea surface temperature (SST) products from this radiometer channels in the 4, 11 and 12 um spectrum are used. The Aqua platform is in a sun synchronous, near polar orbit at 705 km altitude and the MODIS instrument images the entire Earth every 1 to 2 days. The production of the MODIS L2P SST data as part of the Group for High Resolution Sea Surface Temperature (GHRSST) is a joint collaboration between the NASA Jet Propulsion Laboratory (JPL), the NASA Ocean Biology Processing Group (OBPG), and the Rosenstiel School of Marine and Atmospheric Science (RSMAS). RSMAS is responsible for sea surface temperature algorithm development, error statistics and quality flagging, while the OBPG is NASA ground data system for production of daily MODIS ocean products in HDF file format. JPL acquires MODIS ocean granules from the OBPG in near realtime and reformats them to the GHRSST L2P netCDF specification with complete metadata. Data consists of global coverage at 1 km resolution with approximately 300+ five minute observation granules per day. For nighttime data both SST from the 11-12 um, and 4 um bands are available. Ocean chlorophyll a and K490 diffuse attenuation coefficient data are also available for daytime periods as non standard L2P core fields. Note that these ocean color data do not represent the most recent version processed by the OBPG and may not be of research quality.


Reference:
Information on the the MODIS SST algorithm can be found in the following documents, [MODIS Infrared SST Algorithm](http://modis.gsfc.nasa.gov/data/atbd/atbd_mod25.pdf) & [Implementation of SST Processing within the OBPG](http://oceancolor.gsfc.nasa.gov/DOCS/modis_sst/)

# (Measurement noise) $\approx$ (MODIS SST data) - (buoy data) $=$ (bias error) #

$$ d^{obs} = d^{true} + \epsilon $$
$$ C_{d^{obs}} \approx C_N $$
where $C_N$ is the covariance matrix of measurement noise (i.e. bias error), $\epsilon$. Therefore when the linear operator matrix is $M$, we get:
$$ m = Md^{obs} $$
$$ C_m \approx M C_N M^T $$
Computing the equation above may seem equivalent to calculating $ C_m = M C_{d^{obs}} M^T $ but because we want to see how the measurement errors propagate through the system, we should focus on the covariance of the bias error. The spatial covariance of the bias error is:
$$ \sigma^N_{(ij)(kl)} = \frac {1}{N} \Sigma_t
   \big( n(i,j,t)-\overline{n(i,j)} \big) \big( n(k,l,t)-\overline{n(k,l)} \big) $$
where $ \overline {(\hspace{0.3cm})} $ represents the temporal average of each column.

```python
# define dataset location
ddir = '/Volumes/Bucket1/Data/ghrsst/data/L2P/MODIS_A/JPL'
gc = ghrsst_L2P_dataset.GHRSSTCollection(ddir)


def bias_matrix(fname):
    from satdatatools import ghrsst_L2P_dataset, aggregator, utils
    import numpy as np
    try:
        f = ghrsst_L2P_dataset.GHRSSTFile(fname)
    except ValueError:
        return None
    mask = (f.proximity_confidence<5)
    bias = np.ma.masked_array(f.SSES_bias_error, mask)[0]
    dim_lat = f.lat.shape
    dim_lon = f.lon.shape
    if f.lat[0,:]>7 & f.lon[:,0]>157 & f.lat[dim_lat[0],:]<29 & f.lon[:,dim_lon[1]]<-175:
        for m in range(dim_lon[1]):
            if f.lon[0,m] > 157:
                m_west = m
                break
        for m in range(dim_lon[1],-1,-1):
            if f.lon[0,m] <-175:
                m_est = m
                break
        for m in range(dim_lat[0]):
            if f.lat[m,0] > 7:
                m_sud = m
                break
        for m in range(dim_lat[0],-1,-1):
            if f.lat[m,0] < 29:
                m_nord = m
                break
        data = bias[m_sud:m_nord, m_est:m_west]
        lat = f.lat[m_sud:m_nord, m_est:m_west]
        lon = f.lon[m_sud:m_nord, m_est:m_west]
    return data, lat, lon


```
