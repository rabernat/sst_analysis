# Load data #

```python
d = np.load('data/MODIS_L4_SST_QMDA.npz')
print d.keys()
data = d['SST'][500:]
mask_mat = d['mask'][500:]
#error_mat = d['error_mat']
data_size = data.shape
print data_size, mask_mat.shape
```

## Stack seasons together ##

```python
#data_MAM = np.ma.masked_array(data[:,:,60:152], mask_mat[:,:,60:152]!=0)
data_JJA = np.ma.masked_array(data[:,:,152:244], mask_mat[:,:,152:244]!=1)
#data_SON = np.ma.masked_array(data[:,:,244:335], mask_mat[:,:,244:335]!=0)
index_win = np.concatenate((np.arange(335,365), np.arange(0,60)), axis=1)
print index_win
data_DJF = np.ma.masked_array(data[:,:,np.r_[index_win]], mask_mat[:,:,np.r_[index_win]]!=1)
print data_JJA.shape, data_DJF.shape
```

```python
JJA_point = data_JJA[np.round(np.floor(data_size[0]*0.5)),np.round(np.floor(data_size[1]*0.5)),:]
DJF_point = data_DJF[np.round(np.floor(data_size[0]*0.5)),np.round(np.floor(data_size[1]*0.5)),:]
```

```python
JJA_res = JJA_point-JJA_point.mean()
DJF_res = DJF_point-DJF_point.mean()
```

```python
#Unnormalized autocorrelation
#auto_MAM = np.correlate(data_MAM[:], data_MAM[:], mode='full')
#auto_JJA = acorr(JJA_res[:])#, normed = True)
auto_JJA = np.correlate(JJA_res[:], JJA_res[:], mode='full')
#auto_SON = np.correlate(data_SON[:], data_SON[:], mode='full')
#auto_DJF = acorr(DJF_res[:])#, normed = True)
auto_DJF = np.correlate(DJF_res[:], DJF_res[:], mode='full')
print auto_JJA.shape, auto_DJF.shape
```

```python
plt.figure(figsize=(10,5))
plt.xlim(0,100)
plt.title('SST in summer', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('time [days]', fontsize=15)
plt.ylabel('temperature [Celcius]', fontsize=15)
plt.plot(JJA_res)

plt.figure(figsize=(10,5))
plt.xlim(0,100)
plt.title('SST in winter', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('time [days]', fontsize=15)
plt.ylabel('temperature [Celcius]', fontsize=15)
plt.plot(DJF_res)

#plt.figure(figsize=(10,5))
#plt.plot(auto_DJF/DJF_res.var())
#plt.plot(auto_DJF)
#plt.title('Autocorrelation in winter')
#plt.xlim(0,190)
```

```python
x_JJA = np.arange(-91,92)
x_DJF = np.arange(-89,90)

fig = plt.figure(figsize=(10,5))
blue, = plt.plot(x_JJA, auto_JJA/JJA_res.var())
green, = plt.plot(x_DJF, auto_DJF/DJF_res.var())
#plt.plot(auto_JJA)
#plt.xlim(-92, 92)
plt.title('Autocorrelation of SST', fontsize=15)
plt.xlabel('Lag [days]', fontsize=15)
plt.ylabel('Amplitude', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend([blue, green], ['Summer', 'Winter'], loc=2, fontsize=15)
```

# Spectral Analysis in the spatial dimension #

```python
Jany = data[np.round(np.floor(data_size[0]*0.5)),:,15]
Janx = data[:,np.round(np.floor(data_size[0]*0.5)),15]
July = data[np.round(np.floor(data_size[0]*0.5)),:,196]
Julx = data[:,np.round(np.floor(data_size[0]*0.5)),196]

print Janx.shape, Julx.shape, Jany.shape, July.shape
```

```python
fJany, Py_den_Jan = signal.periodogram(Jany-Jany.mean(), fy)
fJuly, Py_den_Jul = signal.periodogram(July-July.mean(), fy)

plt.figure(figsize=(10,7))
blue, = plt.semilogy(fJuly*10**6, Py_den_Jul)
green, = plt.semilogy(fJany*10**6, Py_den_Jan)
plt.title('SST spectral in the meridional direction', fontsize=15)
plt.xlabel('wavenumber $(=1/\lambda) *10^{-6} [m^{-1}]$', fontsize=15)
plt.ylabel('PSD $[^\circ C ^2/m]$', fontsize=15)
plt.ylim([10**1, 10**9])
plt.xticks(fontsize=15)
plt.yticks([10**2,10**4,10**6,10**8], fontsize=15)
plt.legend([blue, green], ['July', 'January'], loc=1, fontsize=15)
```
