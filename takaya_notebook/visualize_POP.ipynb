# January & July # 

```python
Jul_start = 31+28+31+30+31+30

Janx = SST[:31,np.round(SST.shape[1]*0.5),:]
Jany = SST[:31,:,np.round(SST.shape[2]*0.5)]
Julx = SST[Jul_start:Jul_start+31,np.round(SST.shape[1]*0.5),:]
July = SST[Jul_start:Jul_start+31,:,np.round(SST.shape[2]*0.5)]

NtJan, Ny, Nx = SST[:31,:,:].shape
NtJul, Ny, Nx = SST[Jul_start:Jul_start+31,:,:].shape

print Nx, Ny, NtJan, NtJul, Janx.shape, Julx.shape, Jany.shape, July.shape
#print July
```

```python
# Hanning window
windowx = sig.hann(Nx)
windowy = sig.hann(Ny)
```

```python
freqx = fft.fftfreq(Nx, dx)
freqy = fft.fftfreq(Ny, dy)
print freqx.shape, freqy.shape
```

```python
PSD_Janx = np.zeros(Nx)
PSD_Julx = np.zeros(Nx)

for n in range(NtJan):
    # Subtract the mean
    Janx_SST = Janx[n,:]-Janx[n,:].mean()
    # Detrend
    Janx_SST = sig.detrend(Janx_SST)
    # Apply a Hanning window
    Janx_SST *= windowx
    FT_Janx = fft.fft(2*Janx_SST/Nx)
    PSD_Janx += np.real(FT_Janx*np.conj(FT_Janx))
    
for n in range(NtJul):
    # Subtract the mean
    Julx_SST = Julx[n,:]-Julx[n,:].mean()
    # Detrend
    Julx_SST = sig.detrend(Julx_SST)
    # Apply a Hanning window
    Julx_SST *= windowx
    FT_Julx = fft.fft(2*Julx_SST/Nx)
    PSD_Julx += np.real(FT_Julx*np.conj(FT_Julx))
    
PSD_Janx = PSD_Janx/NtJan
PSD_Julx = PSD_Julx/NtJul
```

```python
PSD_Jany = np.zeros(Ny)
PSD_July = np.zeros(Ny)

for n in range(NtJan):
    # Subtract the mean
    Jany_SST = Jany[n,:]-Jany[n,:].mean()
    # Detrend
    Jany_SST = sig.detrend(Jany_SST)
    # Apply a Hanning window
    Jany_SST *= windowy
    FT_Jany = fft.fft(2*Jany_SST/Ny)
    PSD_Jany += np.real(FT_Jany*np.conj(FT_Jany))
    
for n in range(NtJul):
    # Subtract the mean
    July_SST = July[n,:]-July[n,:].mean()
    # Detrend
    July_SST = sig.detrend(July_SST)
    # Apply a Hanning window
    July_SST *= windowy
    FT_July = fft.fft(2*July_SST/Ny)
    PSD_July += np.real(FT_July*np.conj(FT_July))
    
PSD_Jany = PSD_Jany/NtJan
PSD_July = PSD_July/NtJul
```

## Confidence interval ##
$$ p = P\Big( \frac{2m\hat{ \Gamma }(k_j)}{b} \leq \Gamma(k_j) \leq \frac{2m\hat{ \Gamma }(k_j)}{a} \Big) $$
where $ \hat{\Gamma}(k_j) $ is the estimated spectra. $ a $ and $ b $ are the $ (1-p)/2 $ and $ (1+p)/2 $ critical values of the $ \chi^2 (2m) $ distribution and $ m $ is the number of chunks that are added up. The spectra follows the $ \chi^2 $ distribution because $ \hat{\Gamma}^{(n)}(k_j) = \frac{1}{m} \Sigma_{0 \leq n \leq m} {|z^{(n)}(k_j)|}^2 $.
Now the $ p $% confidence range is:
$$ \frac{2m\hat{ \Gamma }(k_j)}{b} \leq \Gamma(k_j) \leq \frac{2m\hat{ \Gamma }(k_j)}{a} $$
$$ \therefore log\Big( \frac{2m}{b} \Big) + log\big( \hat{ \Gamma }(k_j) \big) \leq log\big(\Gamma(k_j)\big) \leq log\Big( \frac{2m}{a} \Big) + log\big( \hat{ \Gamma }(k_j) \big) $$
so the confidence interval is:
$$ 2\sigma_{\Gamma} = log\Big( \frac{2m}{a} \Big) - log\Big( \frac{2m}{b} \Big) = log\Big( \frac{b}{a} \Big) $$

```python
p = 0.95
a = (1-p)*0.5
b = (1+p)*0.5
```

# Structure function #

$$ \langle |T(x+ n\Delta x) - T(x)| \rangle \propto | n\Delta x|^H $$
$$ \langle |T(y+ n\Delta y) - T(y)| \rangle \propto | n\Delta y|^H $$
where the brackets, $ \langle \rangle $ represent an averaging
$$ \therefore \hspace{0.2cm} \Sigma_n^x \equiv \frac {\log{ \big( \big \langle \big|T(x+ n\Delta x) - T(x) \big| \big \rangle \big)}}{\log{ | n\Delta x |}} \propto H , \hspace{0.5cm}
\Sigma_n^y \equiv \frac {\log{ \big( \big \langle \big|T(y+ n\Delta y) - T(y) \big| \big \rangle \big)}}{\log{ | n\Delta y |}} \propto H $$
$$ \therefore \hspace{0.2cm} \Sigma_n \equiv \Sigma_n^x + \Sigma_n^y  \propto H $$

## Compute for the numerators of $ \Sigma_n^x $ & $ \Sigma_n^y $ i.e., ##
## $$ \log{ \big( \big \langle \big|T(x+ n\Delta x) - T(x) \big| \big \rangle \big)} $$ & $$ \log{ \big( \big \langle \big|T(y+ n\Delta y) - T(y) \big| \big \rangle \big)} $$ ##

```python
Hi = np.zeros(6)
Hj = np.zeros(6)
sumcounti = np.zeros(6)
sumcountj = np.zeros(6)
A = f.variables['radius'][:]*0.01

for m in range(6):
#    distj, disti = utils.sphere_distance(f.lon, f.lat)

# [0] stands for the first layer in the first dimension
#    dj, di = utils.sphere_distance(f.lon, f.lat)
#    dlon = np.diff(f.lon, n=m, axis=1)
#    dlat = np.diff(f.lat, n=m, axis=0)

# Difference with 2^m gridpoints in between
    dlon = Tlon_kuroshio[:,2**m:] - Tlon_kuroshio[:,:-2**m]
    dlat = Tlat_kuroshio[2**m:] - Tlat_kuroshio[:-2**m]
#    dlon, dlat = np.meshgrid(dlon[,:], dlat[:])
    disti = np.absolute(A*np.cos(np.radians(
               Tlat_kuroshio[2**m:-2**m,:-2**m]))*np.radians(dlon[2**m:-2**m,:]))
    distj = np.absolute(A*np.radians(dlat[:,2**m:-2**m]))

    mask = (disti > 1100)
#    SST = np.ma.masked_array(f.sea_surface_temperature - 273.15, f.proximity_confidence<5)[0]
    SST_kuroshio = np.ma.masked_array(SST_kuroshio[2**m:-2**m,:-2**m] - 273.15, mask)

# Difference with 2^m gridpoints in between
    dSSTi = np.ma.masked_array(np.absolute(SST_kuroshio[:,2**m:] - SST_kuroshio[:,:-2**m])) # .filled(0.)
    dSSTj = np.ma.masked_array(np.absolute(SST_kuroshio[2**m:] - SST_kuroshio[:-2**m])) # .filled(0.)
    counti = (~dSSTi.mask).astype('i4')
    countj = (~dSSTj.mask).astype('i4')
    #counti = (~mask[:,2**m:]).astype('i4')
    #countj = (~mask[2**m:]).astype('i4')
    sumcounti[m] = np.sum(counti)
    sumcountj[m] = np.sum(countj)
    
#    Hi[m] = np.sum(np.log2(dSSTi))
#    Hj[m] = np.sum(np.log2(dSSTj))
    Hi[m] = np.log2(np.sum(np.absolute(dSSTi))/sumcounti[m])
    Hj[m] = np.log2(np.sum(np.absolute(dSSTj))/sumcountj[m])
    
    print m
```
