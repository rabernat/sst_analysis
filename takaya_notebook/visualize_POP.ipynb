```python
# Hanning window
windowx = sig.hann(Nx)
windowy = sig.hann(Ny)
```

```python
freqx = fft.fftfreq(Nx, dx)
freqy = fft.fftfreq(Ny, dy)
print freqx.shape, freqy.shape
```

# Python FFT equation #
$$ y(j) = \sum_{k=0}^{N-1} x(k) e^{-\frac{2\pi i}{n}kj} = \sum_{k=0}^{N-1} x(k) e^{-i\Delta \omega j k} $$
where $ \omega = \Delta \omega j $ and $ \Delta \omega = \frac{2\pi}{n} $. On the other hand, the infinite Fourier transform is
$$ y(j) = \int x(k) e^{-\frac{2\pi i}{n}kj} dk $$
so for the discrete & infinite Fourier transforms to have the same dimensions, 
$$ y(j) = \int x(k) e^{-\frac{2\pi i}{n}kj} dk \approx \Delta k \  \sum_{k=0}^{N-1}  x(k) e^{-\frac{2\pi i}{n}kj} $$
In our case, this becomes
$$ y(k) = \int T(l) e^{-ikl} dl \approx \Delta l \  \sum_{j=0}^{N-1} T(l_j) e^{-ikl_j} = \Delta l \  FFT \big( T(l) \big) $$
where $ k $ is the wavenumber and $ l $ is the spatial length. Hence, the PSD is calculated as
$$ s^2(k) = \frac{2}{L} {\Delta l}^2 {\bigg| FFT \big( T(l) \big) \bigg| }^2 = \frac{2\Delta l}{N} {\bigg| FFT \big( T(l) \big) \bigg| }^2 $$
with units of $ [^\circ C^2 m] $

```python
PSD_Janx = np.zeros(Nx)
PSD_Julx = np.zeros(Nx)

for n in range(NtJan):
    # Subtract the mean
    Janx_SST = Janx[n,:]-Janx[n,:].mean()
    # Detrend
    Janx_SST = sig.detrend(Janx_SST)
    # Apply a Hanning window
    Janx_SST *= windowx
    FT_Janx = fft.fft(Janx_SST)
    PSD_Janx += 2*dx*np.real(FT_Janx*np.conj(FT_Janx))/Nx
    
for n in range(NtJul):
    # Subtract the mean
    Julx_SST = Julx[n,:]-Julx[n,:].mean()
    # Detrend
    Julx_SST = sig.detrend(Julx_SST)
    # Apply a Hanning window
    Julx_SST *= windowx
    FT_Julx = fft.fft(Julx_SST)
    PSD_Julx += 2*dx*np.real(FT_Julx*np.conj(FT_Julx))/Nx
    
PSD_Janx = PSD_Janx/NtJan
PSD_Julx = PSD_Julx/NtJul
```

```python
PSD_Jany = np.zeros(Ny)
PSD_July = np.zeros(Ny)

for n in range(NtJan):
    # Subtract the mean
    Jany_SST = Jany[n,:]-Jany[n,:].mean()
    # Detrend
    Jany_SST = sig.detrend(Jany_SST)
    # Apply a Hanning window
    Jany_SST *= windowy
    FT_Jany = fft.fft(Jany_SST)
    PSD_Jany += 2*dy*np.real(FT_Jany*np.conj(FT_Jany))/Ny
    
for n in range(NtJul):
    # Subtract the mean
    July_SST = July[n,:]-July[n,:].mean()
    # Detrend
    July_SST = sig.detrend(July_SST)
    # Apply a Hanning window
    July_SST *= windowy
    FT_July = fft.fft(July_SST)
    PSD_July += 2*dy*np.real(FT_July*np.conj(FT_July))/Ny
    
PSD_Jany = PSD_Jany/NtJan
PSD_July = PSD_July/NtJul
```

## Confidence interval ##
$$ p = P\Big( \frac{2m\hat{ \Gamma }(k_j)}{b} \leq \Gamma(k_j) \leq \frac{2m\hat{ \Gamma }(k_j)}{a} \Big) $$
where $ \hat{\Gamma}(k_j) $ is the estimated spectra. $ a $ and $ b $ are the $ (1-p)/2 $ and $ (1+p)/2 $ critical values of the $ \chi^2 (2m) $ distribution and $ m $ is the number of chunks that are added up. The spectra follows the $ \chi^2 $ distribution because $ \hat{\Gamma}^{(n)}(k_j) = \frac{1}{m} \sum_{n=0}^m {\big|z^{(n)}(k_j)\big|}^2 $.
Now the $ p $% confidence range is:
$$ \frac{2m}{b} \leq \frac{\Gamma(k_j)}{\hat{\Gamma}(k_j)} \leq \frac{2m}{a} $$
Written in a log scale, this becomes:
$$ log\Big( \frac{2m}{b} \Big) \leq log\Big(\frac{\Gamma(k_j)}{\hat{\Gamma}(k_j)}\Big) \leq log\Big( \frac{2m}{a} \Big) $$
so the confidence interval is:
$$ \sigma_{log(\Gamma/\hat{\Gamma})} = log\Big( \frac{2m}{a} \Big) - log\Big( \frac{2m}{b} \Big) = log\Big( \frac{b}{a} \Big) = const. $$
Another way of writing this is:
$$ error_{high} = log\Big( \frac{2m}{a} \Big), \ error_{low} = log\Big( \frac{2m}{b} \Big) $$
so to move the error bar around in the log-log dimension, we can muliply the values inside the lograrithm by an arbitrary factor A and we get:
$$ error_{high} = log\Big( A\frac{2m}{a} \Big) = log(A)+log\Big( \frac{2m}{a} \Big) $$
$$ error_{low} = log\Big( A\frac{2m}{b} \Big) = log(A)+log\Big( \frac{2m}{b} \Big) $$

```python
p = 0.95
aJan, bJan = stats.chi2.ppf([(1-p)*0.5, (1+p)*0.5], NtJan)
aJul, bJul = stats.chi2.ppf([(1-p)*0.5, (1+p)*0.5], NtJul)
print aJan, bJan
```

\Sigma_q^y(n) \equiv \frac {\log{ \big( \big \langle \big|T(y+ n\Delta y) - T(y) \big| \big \rangle \big)}}{\log{ | n\Delta y |}} \propto H_q $$
$$ \therefore \hspace{0.2cm} \Sigma_q \equiv \Sigma_q^x + \Sigma_q^y  \propto H_q $$
