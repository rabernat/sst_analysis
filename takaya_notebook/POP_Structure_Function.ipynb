# Second Order Structure Function; q=2 (seasonal) #
$$ \langle {|T(x+ n\Delta x) - T(x)|}^q \rangle \propto | n\Delta x|^{H_q} $$
$$ \langle {|T(y+ n\Delta y) - T(y)|}^q \rangle \propto | n\Delta y|^{H_q} $$
where the brackets, $ \langle \rangle $ represent a spatiotemperal averaging
$$ \therefore \hspace{0.2cm} \Sigma_q^x(n) \equiv \frac {\log{ \big( \big \langle {\big|T(x+ n\Delta x) - T(x) \big|}^q \big \rangle \big)}}{\log{ | n\Delta x |}} \propto H_q , \hspace{0.5cm}
\Sigma_q^y(n) \equiv \frac {\log{ \big( \big \langle {\big|T(y+ n\Delta y) - T(y) \big|}^q \big \rangle \big)}}{\log{ | n\Delta y |}} \propto H_q $$
$$ \therefore \hspace{0.2cm} \Sigma_q \equiv \Sigma_q^x + \Sigma_q^y  \propto H_q $$

```python
Feb_SST = SST[31:59,:,:]
```

```python
Hi = np.zeros((28,8))
Hj = np.zeros((28,8))
sumcounti = np.zeros(8)
sumcountj = np.zeros(8)

for t in range(28):
    for m in range(8):
    # Difference with 2^m gridpoints in between
        dSSTi = np.ma.masked_array((np.absolute(Feb_SST[t,:,2**m:] - Feb_SST[t,:,:-2**m]))**2) # .filled(0.)
        dSSTj = np.ma.masked_array((np.absolute(Feb_SST[t,2**m:] - Feb_SST[t,:-2**m]))**2) # .filled(0.)
        counti = (~dSSTi.mask).astype('i4')
        countj = (~dSSTj.mask).astype('i4')
        #counti = (~mask[:,2**m:]).astype('i4')
        #countj = (~mask[2**m:]).astype('i4')
        sumcounti[m] = np.sum(counti)
        sumcountj[m] = np.sum(countj)
    
#    Hi[m] = np.sum(np.log2(dSSTi))
#    Hj[m] = np.sum(np.log2(dSSTj))
        Hi[t,m] = np.sum(np.absolute(dSSTi))/sumcounti[m]
        Hj[t,m] = np.sum(np.absolute(dSSTj))/sumcountj[m]
    
    print t
```

$$ \frac{\partial T}{\partial x} = \frac{1}{a \cos \theta} \frac{\partial T}{\partial \varphi} $$
$$ \frac{\partial T}{\partial y} = \frac{1}{a} \frac{\partial T}{\partial \theta}$$
