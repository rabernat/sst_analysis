```python
SST_clim_JJA = SST_clim[5:8, :, :]
print SST_clim_JJA.shape

SST_clim_sum = np.nanmean(SST_clim_JJA, axis=0)
SST_clim_sum = np.flipud(np.roll(SST_clim_sum, -180, axis=1))
```

# Longitudinal differential #

$$ \frac{\partial T}{\partial x} 
    = \frac{1}{a \cos \theta} \frac{\partial T}{\partial \varphi} $$
    
# Latitudinal differential #

$$ \frac{\partial T}{\partial y} 
    = \frac{1}{a} \frac{\partial T}{\partial \theta}$$

```python
np.testing.assert_allclose(np.diff(lon), np.diff(lon)[0])
np.testing.assert_allclose(np.diff(lat), np.diff(lat)[0])
dlon = np.radians(np.diff(lon)[0])
dlat = np.radians(np.diff(lat)[0])


daveTdlon = ( np.roll(SST_clim_sum, -1, axis=1) - np.roll(SST_clim_sum, 1, axis=1) ) / (2*dlon)
daveTdx = (gfd.A * np.cos(np.radians(lat)))[:,np.newaxis]**-1 * daveTdlon
print daveTdx.shape

daveTdlat = ( np.roll(SST_clim_sum, -1, axis=0) - np.roll(SST_clim_sum, 1, axis=0) ) / (2*dlat)
daveTdy = gfd.A**-1 * daveTdlat
print daveTdy.shape

SST_clim_grad_sum = daveTdx**2+daveTdy**2
```

# Difference of normalized SST gradient between Summer and Winter #

$$ \delta \frac{\overline{|\nabla T|^2}}{|\nabla \overline{T}|^2} $$

```python
plt.figure(figsize=(15,8))
plt.pcolormesh(lla.lonc, lla.latc, 
               np.log10(np.ma.masked_invalid(grad_win/SST_clim_grad_win)) 
               - np.log10(np.ma.masked_invalid(grad_sum/SST_clim_grad_sum)), cmap = 'RdBu_r')
plt.xlim([-180,180])
plt.ylim([-80,80])
plt.clim(-3,3)
plt.colorbar()
plt.title('Normalized difference of SST gradient (Winter - Summer)')
plt.savefig('Normalized_SST_grad_diff.pdf')
```

# Summary & Discussion #

The SST gradient pattern for $ \delta \frac{\overline{|\nabla T|^2}}{|\nabla \overline{T}|^2} $ and $ \delta |\nabla \overline{T}|^2 $ seem almost identical except for the region of ACC. This indicates that the signal of seasonal variability of SST gradient is smothered by the background gradient of SST climatology. In other words, this normalization method is not good in detecting the seasonal variability of submesoscale structures. As for the region of the ACC, the values of $ \delta \frac{\overline{|\nabla T|^2}}{|\nabla \overline{T}|^2} $ and $ \delta |\nabla \overline{T}|^2 $ seem to have opposite signs.

In order to detect the seasonal variability, therefore, we need to consider the coarsening of the data and see whether this affects the maginitude of SST gradients in different seasons.
