```python
import numpy as np
import os
from scipy import stats
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
base_dir = '/Volumes/Bucket1/Data/hybrid_v5_rel04_BC5_ne120_t12_pop62/'
fname = base_dir + 'hybrid_v5_rel04_BC5_ne120_t12_pop62.pop.h.nday1.0046-02-01.nc'
```

```python
p = popanalysis.POPFile(fname)
```

```python
T = p.nc.variables['SST']
```

```python
Nt, Nx, Ny, k, l, PSD_sum = p.power_spectrum_2d(T)
```

```python
plt.figure(figsize=(10,6))
plt.pcolormesh(k,l,np.log10(PSD_sum/Nt))
plt.colorbar()
```

```python
PSD_sum = PSD_sum/Nt
PSD_Febx = PSD_sum.mean(axis=0)/Nx
PSD_Feby = PSD_sum.mean(axis=1)/Ny
```

```python
p = 0.95

a_Febx, b_Febx = stats.chi2.ppf([(1-p)*0.5, (1+p)*0.5], Nt*Nx)
a_Feby, b_Feby = stats.chi2.ppf([(1-p)*0.5, (1+p)*0.5], Nt*Ny)
err_Febx = 2*Nt*Nx*np.array([1/a_Febx, 1/b_Febx])
err_Feby = 2*Nt*Ny*np.array([1/a_Feby, 1/b_Feby])
```

```python
xFeb = 1.5e-5
yFeb = 3e-5
#yJan = 5e0
#yJul = 5e0
A = 1e2

plt.figure(figsize=(10,7))
blue, = plt.loglog(k, PSD_Febx)
green, = plt.loglog(l, PSD_Feby)
magenta, = plt.loglog([xFeb,xFeb], A*err_Febx, 'm', basex=10, basey=10)
red, = plt.loglog([yFeb,yFeb], A*err_Feby, 'r', basex=10, basey=10) 
plt.title('SST power spectral in February', fontsize=15)
plt.legend([blue, green, magenta, red], ['Meridional', 'Zonal', 'Confidence Interval (Meridional)', 'Confidence Interval (Zonal)'], loc=3, fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)


#plt.figure(figsize=(10,7))
#blue, = plt.loglog(k, PSD_Augx)
#green, = plt.loglog(k, PSD_Febx)
#magenta, = plt.loglog([xAug,xAug], A*err_Augx, 'm', basex=10, basey=10)
#red, = plt.loglog([xFeb,xFeb], A*err_Febx, 'r', basex=10, basey=10) 
#plt.title('SST power spectral in the zonal direction', fontsize=15)
#plt.xlabel('wavenumber $(=k/2\pi) [m^{-1}]$', fontsize=15)
#plt.ylabel('PSD $[^\circ C ^2 m]$', fontsize=15)
#plt.xlim([10**(-7),10**(-3)])
#plt.ylim([10**(-8), 10**1])
#plt.xticks(fontsize=15)
#plt.yticks(fontsize=15)
#plt.xticks([10**(-7),10**(-6),10**(-5),10**(-4),10**(-3)],fontsize=15)
#plt.yticks([10**(-7),10**(-6),10**(-5),10**(-4),10**(-3),10**(-2),10**(-1),10**0], fontsize=15)
#plt.legend([blue, green], ['August', 'February'], loc=1, fontsize=15)
#plt.legend([blue, green, magenta, red], ['August', 'February', 'Confidence Interval (August)', 'Confidence Interval (February)'], loc=3, fontsize=15)
#plt.savefig(os.path.join('QMDA_figs','POP_Zonal_AugFeb_spectral.png'))

#xFeb = 5e-5
#xAug = 2e-5
#A = 1e4
#plt.figure(figsize=(10,7))
#blue, = plt.loglog(l, PSD_Augy)
#green, = plt.loglog(l, PSD_Feby)
#magenta, = plt.loglog([xAug,xAug], A*err_Augy, 'm', basex=10, basey=10)
#red, = plt.loglog([xFeb,xFeb], A*err_Feby, 'r', basex=10, basey=10) 
#plt.title('SST power spectral in the meridional direction', fontsize=15)
#plt.xlabel('wavenumber $(=k/2\pi) [m^{-1}]$', fontsize=15)
#plt.ylabel('PSD $[^\circ C ^2 m]$', fontsize=15)
#plt.xlim([10**(-7),10**(-3)])
#plt.ylim([10**(-8), 10**1])
#plt.xticks(fontsize=15)
#plt.yticks(fontsize=15)
#plt.xticks([10**(-7),10**(-6),10**(-5),10**(-4),10**(-3)],fontsize=15)
#plt.yticks([10**(-7),10**(-6),10**(-5),10**(-4),10**(-3),10**(-2),10**(-1),10**0], fontsize=15)
#plt.legend([blue, green], ['August', 'February'], loc=1, fontsize=15)
#plt.legend([blue, green, magenta, red], ['August', 'February', 'Confidence Interval (August)', 'Confidence Interval (February)'], loc=3, fontsize=15)
#plt.savefig(os.path.join('QMDA_figs','POP_Meridional_AugFeb_spectral.png'))
```
